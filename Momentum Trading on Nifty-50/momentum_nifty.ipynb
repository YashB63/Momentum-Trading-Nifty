```python
import yfinance as yf
import numpy as np
import pandas as pd
```

```python
stocks = pd.read_html('https://en.wikipedia.org/wiki/NIFTY_50')[2].Symbol
```

```python
stocks.head(5)
```

```python
stocks = stocks + '.NS'
```

```python
stocks = stocks.to_list()
```

```python
df = yf.download(stocks, start='2015-1-1')['Close']
```

```python
df
```

```python
returns_df = df.pct_change()
```

```python
returns_df
```

```python
monthly_returns = (returns_df + 1).resample('M').prod()
```
